# Check F-score of the decoding results

```python
import pickle
import pandas as pd
from sklearn.metrics import f1_score

from param import *

output_dir = ParamDir().output_dir
datalist = ParamDir().data_path_list
```

```python
f1_results = []
for data_dir in datalist:
    data_name = str(data_dir).split('/')[-1]
    mouse_type = "KO" if "KO" in data_name else "WT"
    # with open(ParamDir().output_dir/data_name/"tsc_shuffle_SVM_segment label shuffling_train_rocket.pickle","rb") as f:
    #     shuffle_results = pickle.load(f)
    # shuffle_scores = []
    # for shuffle_r in shuffle_results:
    #     clf = shuffle_r["estimator"]
    #     (X_train, y_train), (X_test, y_test) = shuffle_r["data"]
    #     shuffle_score = f1_score(clf.predict(X_test), y_test, average='macro')
    #     shuffle_scores.append(shuffle_score)

    with open(ParamDir().output_dir/data_name/"tsc_train_rocket_SVM_threshold_segment_False.pickle","rb") as f:
        original_results = pickle.load(f)
    clf = original_results["estimator"]
    (X_train, y_train), (X_test, y_test) = original_results["data"]
    real_score = f1_score(clf.predict(X_test), y_test, average=None)
    real_score_average = f1_score(clf.predict(X_test), y_test, average="micro")

    f1_results.append(list(real_score)+[real_score_average]+[data_name]+[mouse_type])
    
```

```python
df_f1_score = pd.DataFrame(f1_results, columns=["1", "2", "3", "4", "averaged score", "data name", "mouse type"])
```
