# Check mice mobility from raw data

```python
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
from tqdm import tqdm

from dataloader import BaseDataset
from param import *

sns.set_theme(style="white")
```

# Representatives

```python
WT_path = ParamDir().data_list[2]

fig, axes = plt.subplots(nrows=2, figsize=(7,5), tight_layout=True, sharex=True)

WT_dataset = BaseDataset(WT_path, ParamData().mobility, ParamData().shuffle, ParamData().random_state)

data_len = len(WT_dataset.coords_xy)
line1 = axes[0].plot(WT_dataset.coords_xy[range(0, data_len, 3), 0]/5,label="x-axis", lw=1, c="r", alpha=0.7)
line2 = axes[0].plot(WT_dataset.coords_xy[range(0, data_len, 3), 1]/5,label="y-axis", lw=1, c="black", alpha=0.5)
axes[0].set_ylabel("coordinates(cm)")

yy = WT_dataset._discretize_coords()[range(0, data_len, 3)]
xx = range(len(yy))
line3 = axes[1].plot(xx ,yy,label="position", lw=1, c="black", marker=".", alpha=.5, markersize=3)
axes[1].set_ylabel("position")
axes[1].set_xlabel("time(s)")
axes[1].set_ylim(.5, 4.5)

lines = [line1, line2, line3]
axes[1].legend(line1 + line2 + line3, [i[0].get_label() for i in lines],bbox_to_anchor=(1,-.4), ncol=3)

plt.suptitle("coordinates and position")
```
